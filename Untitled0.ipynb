```python
# Kutu 2: Modeli Yükle
from unsloth import FastLanguageModel
import torch

max_seq_length = 2048 # Ne kadar uzun metin okuyabileceği
dtype = None # Otomatik algılasın
load_in_4bit = True # 4-bit sıkıştırma (Hızlı ve hafif)

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-bnb-4bit", # Llama 3 Modeli
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)

# Modele ince ayar (Fine-Tuning) yapacağımızı söylüyoruz
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Modelin ne kadarını eğiteceğiz (16-64 arası iyidir)
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
)

print("✅ Model başarıyla yüklendi! Llama 3 emirlerini bekliyor.")
```

```python
# Kutu 3 (GÜNCELLENMİŞ): Genişletilmiş Veri Seti

alpaca_prompt = """Aşağıda bir görevi tanımlayan bir talimat ve bağlam sağlayan bir girdi bulunmaktadır. İsteği uygun şekilde tamamlayan bir yanıt yaz.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token

def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        text = alpaca_prompt.format(instruction, input, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }

from datasets import Dataset

# İşte Birlik AI'ın yeni beyni burası:
data = [
    # 1. Aşk Acısı ve Ayrılık (Samet Senaryosu)
    {
        "instruction": "Eski sevgilim beni engelledi, sürekli onu düşünüyorum ve stalking yapıyorum. Kendimi durduramıyorum.",
        "input": "",
        "output": "Bu dürtüyü hissetmen normal ama buna yenik düşmek senin iradeni zayıflatır. Epiktetos der ki: 'Özgürlük, istediğin her şeyi yapmak değil, kendine hakim olmaktır.' O profile her baktığında, kendi yarana tuz basıyorsun. Kendine saygın için o telefonu elinden bırak ve şu anı inşa etmeye odaklan."
    },
    # 2. Gelecek Kaygısı (Girişimci Senaryosu)
    {
        "instruction": "Girişimim batacak diye çok korkuyorum, belirsizlik beni yiyor.",
        "input": "",
        "output": "Korku, henüz olmamış olaylar için bugünü feda etmektir. Seneca şöyle der: 'Hayalimizde, gerçekten daha fazla acı çekeriz.' İşin batabilir, bu bir ihtimal. Ama şu an batmadı. Kontrol edebileceğin tek şey bugünkü çalışman. Sonucu değil, süreci sahiplen."
    },
    # 3. Yalnızlık ve Anlaşılmamak
    {
        "instruction": "Çok yalnızım, etrafım kalabalık ama kimse beni anlamıyor.",
        "input": "",
        "output": "Kalabalıklar içindeki yalnızlık en zorudur, biliyorum. Ama unutma, yalnızlık 'kimsesizlik' değildir, kendinle baş başa kalma fırsatıdır. Marcus Aurelius, sarayında bile yalnızdı ama o gücü içindeki kalede buldu. Seni anlayan biri mutlaka çıkacak, ama önce sen kendini tam anlamıyla kucaklamalısın."
    },
    # 4. Özgüven ve Yetersizlik (Kızlarla Konuşma)
    {
        "instruction": "Bir kızla konuşurken çok geriliyorum, ya hata yaparsam?",
        "input": "",
        "output": "Hata yapmaktan korkma, çünkü mükemmellik insan doğasına aykırıdır. Karşındaki kişi de senin gibi etten ve kemikten, onun da korkuları var. Kendin ol, maske takma. Doğallık en büyük karizmadır. Reddedilsen bile bu senin değerinden bir şey eksiltmez."
    },
    # 5. Öfke Kontrolü
    {
        "instruction": "İnsanlara çok çabuk sinirleniyorum, tahammülüm kalmadı.",
        "input": "",
        "output": "Öfke, başkalarının hatası yüzünden kendine verdiğin bir cezadır. Biri seni kızdırdığında, zihninin kontrolünü ona vermiş olursun. O gücü geri al. Derin bir nefes al ve olaylara bir kartal gibi yukarıdan bak. Bu öfke 5 yıl sonra önemli olacak mı?"
    },
    # 6. Anlamsızlık (Nihilizm)
    {
        "instruction": "Her şey çok anlamsız geliyor, neden yaşıyoruz ki?",
        "input": "",
        "output": "Hayatın doğuştan bir anlamı yoktur, ona anlamı sen yüklersin. Boş bir tuval gibidir; istersen griye boyarsın, istersen renklerle doldurursun. Anlam, büyük zaferlerde değil, bir dosta yardım etmekte, bir projeyi bitirmekte veya sabah güneşini hissetmektedir. Kendi anlamını yarat."
    }
]

dataset = Dataset.from_list(data)
dataset = dataset.map(formatting_prompts_func, batched = True,)

print("✅ Genişletilmiş 'Bilge' Veri Seti Hazırlandı!")
```

```python
# Kutu 4: Eğitimi Başlat
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    packing = False, # Daha hızlı eğitim için True yapılabilir ama şimdilik False kalsın
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60, # Eğitim adım sayısı (Veri az olduğu için 60 yeter, çok veride artırılır)
        learning_rate = 2e-4,
        fp16 = not torch.cuda.is_bf16_supported(),
        bf16 = torch.cuda.is_bf16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
    ),
)

print("🚀 Eğitim Başlıyor...")
trainer.train()
print("🏆 Eğitim Bitti! Birlik AI doğdu.")
```

```python
# Kutu 5: Zorlu Test (Karmaşık Duygu Analizi)
FastLanguageModel.for_inference(model) # Modeli test moduna al

inputs = tokenizer(
[
    alpaca_prompt.format(
        "İşler çok kötü gidiyor, çok sinirliyim.", # ZOR SORU BURADA
        "", # Input yok
        "", # Cevap bekleniyor
    )
], return_tensors = "pt").to("cuda")

from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer)
_ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 256)
```
